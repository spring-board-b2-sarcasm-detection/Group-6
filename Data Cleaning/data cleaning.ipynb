## step1 - importing the dataset

```python
import pandas as pd
df = pd.read_excel('Movie sarcasm Final Dataset.xlsx')
df
```

## step2 - removing duplicate reviews

```python
df = df.drop_duplicates(subset='Review' , keep ='first')
df
```

## step 3: converting to lower case

coverting all the Uppercase to lowercase using str.lower 

```python
df['cleaned_data'] = df['Review'].str.lower()
df.cleaned_data
```

## step 4 - Punctuation removal
all the punctuation marks like ' " ; ? . ! are removed

```python
import string
# Function to remove punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

# Apply the function to the 'cleaned_data' column
df['cleaned_data'] = df['cleaned_data'].apply(remove_punctuation)
df.cleaned_data
```

## step 5 removal of stopwords

```python
from nltk.corpus import stopwords

# Define English stopwords
stop_words = set(stopwords.words('english'))

# Function to remove stopwords from a sentence
def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

# Apply the function cleaned_data
df['cleaned_data'] = df['cleaned_data'].apply(remove_stopwords)


df.cleaned_data
```

## updating the new data to the excel sheet

```python
df_new = pd.read_excel('Cleaned Dataset.xlsx')
df_new['cleaned_data'] = df['cleaned_data']
df_new['Sentiment'] = df['Sentiment']
df_new['Sarcasm'] = df['Sarcasm']
df_new.to_excel('Cleaned Dataset.xlsx', index=False)
df_new
```
